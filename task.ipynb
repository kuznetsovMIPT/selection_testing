# Тестовое задание

##  Идея
Необходимо написать функцию, позволяющую вычислять характеристическую скорость, необходимую для смены орбиты.

## Входные данные
На вход необходимо принять текующую орбиту объекта, целевую орбиту объекта (подумать, в каком формате), гравитационный параметр Земли. Достаточно рассмотреть лишь эллиптические орбиты.

## Результат работы
В качестве результата работы функция должна возвращать необходимую характеристическую скорость.

## Требование к коду
Функция должна быть реализована в виде статической библиотеки. Сборка библиотеки должна осуществляться при попощи Cmake.

## Документация
Код должен быть задокументирован. Помимо док-стрингов должен быть создан файл Readme, в котором описаны интерфесы функции, описание работы функции, ссылки на источники информации.

## Тесты
Необходимо провести тестирование функции с использованием средств gtest. Тестирование должно быть понятным (тесты также стоит описать) и глобальным (тесты должны покрывать весь код). Любой тест должен заканчиваться вызовом ASSERT.

## Репозиторий
Код, тесты и документацию следует разместить в своем ЗАКРЫТОМ репзитории github. Пригласить https://github.com/kuznetsovMIPT и написать на почту kuznetsov.aa@mipt.ru письмо с ФИО, номером группы и ссылкой на репозиторий.
